```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')

# ignore warning messages
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# create a 10 X 10 matrix, with values between -1 and 1
matrix = np.random.uniform(-1, 1, size=(10,10))

# plot the heatmap
plt.figure(figsize=(10,10))
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, annot = True)
plt.show()
```

```python
# create a 1000 X 3 matrix, with values between -1 and 1
nrow = 1000
ncol = 3
matrix = np.random.uniform(-1, 1, size=(nrow,ncol))

# randomly tags each row data with 0, 1, and 2
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# plot PairPlot (with upper scatter, diagonal histogram, and lower density plot)
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

```python
# create a 1000 X 3 matrix, with values are normal distribution
nrow = 1000
ncol = 3
matrix = np.random.normal(size=(nrow,ncol))

# randomly tags each row data with 0, 1, and 2
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# plot PairPlot (with upper scatter, diagonal histogram, and lower density plot)
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])

grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
