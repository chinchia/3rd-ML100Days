```python
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# generate 1000 x with range 0 ~ 50 randomly
x = np.random.randint(0, 50, 1000)

# make x and y negatively correlated, and add some noise
y = 100 - x + np.random.normal(0, 5, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```
